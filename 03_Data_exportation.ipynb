```python
import pandas as pd
from fpdf import FPDF
from fpdf.enums import VAlign, Align, XPos, YPos
```

```python
chinese_spanish_vocabulary = pd.read_parquet("data/chinese_spanish_vocabulary.parquet").sort_values('index', ascending=True)
```

```python
# transform index into string in order to avoid encoding errors.
chinese_spanish_vocabulary['index'] = chinese_spanish_vocabulary['index'].astype('string')
```

```python


class PDF(FPDF):
    def header(self):
        if self.page_no() == 1:
            self.set_font("chinese_font", "B", 18)
            
            self.cell(
                w=0, 
                h=10,
                text="Vocabulario HSK-2",
                align=Align.C, 
                new_x=XPos.LMARGIN, 
                new_y=YPos.NEXT 
            )
            self.ln(10)
    def footer(self):        
        self.set_y(-15) 
        
        self.set_font("chinese_font", "B", 6) 
        
        page_text = f"Pagina {self.page_no()} / {{nb}}"
        
        self.cell(0, 10, "github.com/Mortadelo24/hsk-vocabulary", align=Align.C)




pdf = PDF()
pdf.add_font("chinese_font", "", 'fonts/Noto_Sans_SC/static/NotoSansSC-Regular.ttf')
pdf.add_font("chinese_font", "B", 'fonts/Noto_Sans_SC/static/NotoSansSC-Bold.ttf')
pdf.add_page()

pdf.set_font("chinese_font", size=12)

with pdf.table(
    col_widths=(20, 40, 60, 130), 
    text_align="LEFT",
    borders_layout="SINGLE_TOP_LINE",
    v_align=VAlign.T,
    repeat_headings=0, 
    ) as table:
    title_row = table.row()
    title_info = ("ID", "Characteres", "Pinyin", "Definicion")
    
    for x in title_info:
        title_row.cell(x, padding=(2, 1))

    for data_row in chinese_spanish_vocabulary.itertuples(index=False):
        row = table.row()

        for x in data_row:
            row.cell(x, padding=(2, 1))
  

pdf.output("output/hsk2_vocabulario.pdf")
```
